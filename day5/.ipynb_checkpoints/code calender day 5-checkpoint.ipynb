# Sorting one container at a time

```python
import re

def brute_force_sorting_one_container(statement_list, container_stacks):
    for statement in statement_list:
        # create data for crane to move the containers
        cut_statement = re.sub('\D', '',statement)
        if len(cut_statement) == 4:
            number_containers = int(cut_statement[0:2])
            source_stack = int(cut_statement[2]) - 1
            target_stack = int(cut_statement[3]) - 1
        else:
            number_containers = int(cut_statement[0])
            source_stack = int(cut_statement[1]) - 1
            target_stack = int(cut_statement[2]) - 1
            
        #print(number_containers, source_stack, target_stack)
        #print(range(number_containers))
        # change container positions
        for i in range(number_containers):
            container = container_stacks[source_stack].pop()
            container_stacks[target_stack].append(container)
            
    return container_stacks
```

# move 3 containers if needed

```python
import re

def brute_force_sorting_multiple_container(statement_list, container_stacks):
    for statement in statement_list:
        # create data for crane to move the containers
        cut_statement = re.sub('\D', '',statement)
        if len(cut_statement) == 4:
            number_containers = int(cut_statement[0:2])
            source_stack = int(cut_statement[2]) - 1
            target_stack = int(cut_statement[3]) - 1
        else:
            number_containers = int(cut_statement[0])
            source_stack = int(cut_statement[1]) - 1
            target_stack = int(cut_statement[2]) - 1
            
        #print(number_containers, source_stack, target_stack)
        #print(range(number_containers))
        # change container positions
        container = container_stacks[source_stack][-number_containers:]
        del container_stacks[source_stack][-number_containers:]
        container_stacks[target_stack].extend(container)
            
    return container_stacks
```
