```python
def calculate_overlap(pairgen):
    
    overlaps = 0
    
    for pair in pairgen:
        range1, range2 = pair.split(',')
        range1 = range1.split('-')
        range2 = range2.split('-')

        print('pair:', range1, range2)

        if (range1[0] <= range2[0] and range1[1] >= range2[1]) or (range2[0] <= range1[0] and range2[1] >= range1[1]):
            print('overlapping:', range1, range2)
            overlaps += 1
            
    return overlaps
```

```python
pairgen = (pair for pair in get_puzzle_input().split('\n'))
calculate_overlap(pairgen)
```
