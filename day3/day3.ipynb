```python
def get_shared_items(rucksacks):
    rucksacks = rucksacks.split('\n')
    
    shared_items = []
    
    for rucksack in rucksacks:
        split = int(len(rucksack)/2)
        comp1, comp2 = rucksack[0:split], rucksack[-split:]
        
        for letter in comp1:
            if letter in comp2:
                #print(comp1, comp2, letter)
                shared_items.append(letter)
                break

    return shared_items
```

```python
get_shared_items(get_puzzle_input())
```

```python
import string

def calculate_priority(shared_items):
    lowercase = dict(zip(string.ascii_lowercase, range(1,27)))
    uppercase = dict(zip(string.ascii_uppercase, range(27,53)))
    
    amount = 0
    
    for item in shared_items:
        if item in lowercase.keys():
            #print(item, lowercase[item])
            amount += lowercase[item]
        else: 
            #print(item, uppercase[item])
            amount += uppercase[item]
            
    return amount
```

```python
calculate_priority(get_shared_items(get_puzzle_input()))
```

```python
def get_group(elves):
    return elves.split('\n')[0:3]
    
```

```python
def get_shared_item_by_group(elves):
    rucksacks = elves.split('\n')
    group_badges = []
    
    # list contains one empty element, that's why len > 2
    while rucksacks != [] and len(rucksacks) > 2:
        g1, g2, g3 = rucksacks[0:3]
        del rucksacks[0:3]
        
        for letter in g1:
            if letter in g2 and letter in g3:
                group_badges.append(letter)
                break
    return group_badges
    
```
